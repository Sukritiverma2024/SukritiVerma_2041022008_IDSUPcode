```python
import random as rd
import pandas as pd
import numpy as np
import collections
a=[]
for i in range(60,71):
    a.append(i)
print(a)
#This code will show all the possible values that can be choosen between 60 to 71.
```

```python
Ages=[]
for i  in range(1000):
    nm=(rd.choice(a))
    Ages.append(nm)
    
print(Ages)
# This code randomly selects 1000 nos. in the given range (Includes repition due to restraining boundary of a's range.)
```

```python
mean=np.mean(Ages)
print("mean Ages, μ=",mean)
#Calculates and prints the mean of Ages(the randomly generated 1000 values between 60 to 71.)
```

```python
stand_dev=np.std(Ages)
print("Standard deviation of Ages, δ=",stand_dev)
#Calculates and shows the standard deviation of Ages.
```

```python
import matplotlib.pyplot as plt
plt.hist(Ages)
#Ploting the histogram of above data ages
```

```python
sample=[]
Smean=[]
for j in range(500):
    for i in range(30):
        j=rd.choice(Ages)
        #randomly choosing 30 nos. from ages
        sample.append(j)
    s=np.mean(sample)
    Smean.append(s)
print(Smean)
#Calculating the mean of sample and storing it in smean.
```

```python
plt.hist(Smean)
#ploting the histogram of smean.
```

```python
print("mean of S.mean, μ1 = ",np.mean(Smean))
#Calculating and printing Mean of S.mean.
```

```python
print('Standard Deviation od S.mean, δ1 = ',np.std(Smean))
#Calculating and printing Standard Deviation of S.mean.
```
